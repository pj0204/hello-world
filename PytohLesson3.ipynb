# Regresión Lineal
Dado una serie de datos X e Y, hallaremos la ecuación de la recta que mejor se ajusta. La recta es de la forma $y=a+bx$.




| x | y |
|---|---|
| 1 | 3 |
| 2 | 5 |
| 3 | 7 |
| 4 | 9 |

Recordemos que :



> a=$\frac{\sum y \sum x^2 - \sum x \sum x y}{n \sum x^2 -(\sum x)^2}$



> b=$\frac{n \sum xy -\sum x \sum y}{n \sum x^2 -(\sum x)^2}$ 



````
import numpy as np #importando libreria numpy
x=np.array([1,2,3,4]) #data
y=np.array([3,5,7,9]) #data

sumY=sum(y)
#print(sumY)
sumX2=sum(x**2)
#print(sumX2)
sumX=sum(x)
#print(sumX)
sumXY=sum(x*y)
#print(sumXY)
n=len(x)

a=(sumY*sumX2-sumX*sumXY)/(n*sumX2-sumX**2)
print('a: ',a)

b=(n*sumXY-sumX*sumY)/(n*sumX2-sumX**2)
print('b: ',b)

print('Ecuacion de la recta')
print('y=',a,'+',b,'x')
````

Lo mejor sería tener una visualización de lo ocurrido.


````
import matplotlib.pyplot as plt

plt.scatter(x, y,c='blue',alpha=0.5)
plt.title('Mi grafico')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.plot(x,y)
plt.show()
````
